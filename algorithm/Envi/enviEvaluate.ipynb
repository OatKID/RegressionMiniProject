```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1p1 = pd.read_csv('../../Algorithm/Envi/1_predict_1/evaluate.csv')
df_1p1
```

```python
df_sp1 = pd.read_csv('../../Algorithm/Envi/sequence_predict_1/evaluate.csv')
df_sp1
```

R2 Evaluate

```python
baseLine_1p1 = df_1p1.loc[0]
gradient_1p1 = df_1p1.loc[1]
linear_1p1 = df_1p1.loc[2]

baseLine_sp1 = df_sp1.loc[0]
gradient_sp1 = df_sp1.loc[1]
linear_sp1 = df_sp1.loc[2]
```

```python
x_1p1 = np.array(["BaseLine","Linear","GradientBoosting"])
y_1p1 = np.array([baseLine_1p1['R2'],linear_1p1['R2'],gradient_1p1['R2']])

y_sp1 = np.array([baseLine_sp1['R2'],linear_sp1['R2'],gradient_sp1['R2']])

X_axis = np.arange(len(x_1p1)) 


plt.figure(figsize=(8,6))
plt.bar(X_axis - 0.15,y_1p1, width=0.3, label='1 Predict 1', edgecolor = 'black')
plt.bar(X_axis + 0.15 ,y_sp1, width=0.3, label='Sequence Predict 1', edgecolor = 'black')

plt.ylabel("R2 (R-Squere) %")
plt.xlabel("Regression Models")
plt.title("R2 Score From Envi Dataset")
plt.xticks(X_axis, x_1p1)

for index,value in enumerate(y_1p1):
    plt.text(x=index - 0.05, y=value + 0.01, s=str(f"{value:.4f}"),ha='right')

for index,value in enumerate(y_sp1):
    plt.text(x=index + 0.05, y=value + 0.01, s=str(f"{value:.4f}"),ha='left')

plt.legend()
plt.show()
```

MAE Evaluate

```python
x_1p1 = np.array(["BaseLine","Linear","GradientBoosting"])
y_1p1 = np.array([baseLine_1p1['Mae'],linear_1p1['Mae'],gradient_1p1['Mae']])

y_sp1 = np.array([baseLine_sp1['Mae'],linear_sp1['Mae'],gradient_sp1['Mae']])

X_axis = np.arange(len(x_1p1)) 


plt.figure(figsize=(8,6))
plt.bar(X_axis - 0.15,y_1p1, width=0.3, label='1 Predict 1', edgecolor = 'black')
plt.bar(X_axis + 0.15 ,y_sp1, width=0.3, label='Sequence Predict 1', edgecolor = 'black')

plt.ylabel("MAE (Mean Absolute Error)")
plt.xlabel("Regression Models")
plt.title("MAE Score From Envi Dataset")
plt.xticks(X_axis, x_1p1)

for index,value in enumerate(y_1p1):
    plt.text(x=index - 0.01, y=value + 0.1, s=str(f"{value:.2f}"),ha='right')

for index,value in enumerate(y_sp1):
    plt.text(x=index + 0.01, y=value + 0.1, s=str(f"{value:.2f}"),ha='left')

plt.legend()
plt.show()
```
